SOLUTION:

This follows a source free RLC circuit with interval (4 <= t <= 21) where through derivation via KCL @ top node, we get:

V" + (1/RC)V' + (1/LC)V = 0, where values of R, L, C are given above.

with Initial values (@ t = 4):

V(4) = 10V

IL(4) = 2A

and V_L = V_R = V_C = V, since it's a parallel RLC Circuit (@ any time interval).

However in a parallel circuit, current in each parallel branch aren't equal, therefore then we KCL (@ t = 4):

0 = I_R(4) + V(4)/R + C(dV/dt)

By solving the 2nd order differential equation, we can get the following damping factors: 

S1 = -1

S2 = -2 


Then we get the following general solution of a homogeneous 2nd order DE(Overdamped):

FOR V(t):
$$V\!\left(t\right)= C1exp(-t) + C2exp(-2t) $$

Then,

FOR f(t):
$$f\!\left(t\right) = V^{\prime}\!\left( t \right)  = -C1exp(-t) -2C2exp(-2t)= 0 $$


Where value of f(t) is set @ minimum, to verify, its 2nd derivative should be > 0.

Values of C1 and C2 upon solving the differential equation:

C1 = -764.374(approximate of its exponential value) = -14e^4

C2 = 71542.992(approximate of its exponential value) = 24e^8


````python
C = 500 #Millifarad
R = 2/3 #Ohm
L = 1000 #Millihenry

S_1, S_2 = -1, -2 #damping factors
C_1, C_2 = -14*mt.exp(4), 24*mt.exp(8)

def V_L(t): #V_L(t) = V(t) across the circuit
    V = C_1*mt.exp(S_1*t) + C_2*mt.exp(S_2*t) 
    return V

def resid_from_x(t): # V'(t) = f(t), where it is @ minimum. 
    V_prime = S_1*C_1*mt.exp(S_1*t) + S_2*C_2*mt.exp(S_2*t)
    return V_prime
````

Use the Newton-Raphson method to find the root of $f\!\left(t\right)$.

Define the Python variable `T0` that will store the initial estimate of the root.
Run
[`scipy.optimize.newton()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.newton.html)
such that you get the (approximate) root as well as convergence information,
and store the said outputs in Python variables `t_sc` and `t_sc_info`, respectively.
You may provide non-default values to the tolerances and to the iteration budget.

In the succeeding code cell,
replace the `pass` statement with your code.

````python
T0 = 1
iteration = 100
epsilon = 1e-8

# newton search without derivative is essentially secant method
t_sc, t_sc_info = spo.newton(
    resid_from_x, T0,
    tol = epsilon, maxiter = iteration,
    full_output = True,
    disp = False
)

print(f"Finding a root starting at {T0}:")
````
