## Modelling

Derive a system of eight (8) linear algebraic equations
in terms of the said eight desired quantities.
Arrange the equations in the matrix-vector form
$\boldsymbol{A} \boldsymbol{x} = \boldsymbol{b}$,
where $\boldsymbol{x}$ contains the unknowns
in the order presented above.

Define two Python functions, `get_coeffs()` and `get_rhs()`.

`get_coeffs()` takes two positional arguments in the following order:

- `b`, representing the dimension $b$, and
- `h`, representing the dimension $h$,

and returns a NumPy array representing the coefficient matrix $\boldsymbol{A}$.

On the other hand,
`get_rhs()` takes two positional arguments in the following order:

- `Pv`, representing $P_{\text{v}}$, and
- `Ph`, representing $P_{\text{h}}$,

and returns a NumPy array representing the right-hand-side vector $\boldsymbol{b}$.

```python
def get_coeffs(b, h) :
    
    b = 10 # assumed value 
    h = 15 # assumed value

    # dimensions (b & h) are assumed to be any number
    
    def unit_vector (x_hat, y_hat): # Triangle dimensions
        x_hat = 0.5*b/(mt.sqrt(mt.pow(h,2) + mt.pow(0.5*b,2)))
        y_hat = h/(mt.sqrt(mt.pow(h,2) + mt.pow(0.5*b, 2)))

        A = np.array([
            [h, b, 0, 0, 0, 0, 0, 0],       # Equation 1
            [0, 0, -b, 0, 0, 0, 0, 0],      # Equation 2
            [0, 0, 1, y_hat, 0, 0, 0, 0],   # Equation 3
            [0, 0, 0, x_hat, -1, 0, 0, 0],  # Equation 4
            [0, 1, 0, 0, 0, 0, -1, 0],      # Equation 5
            [-1, 0, 0, 0, 0, 0, 0, -1],     # Equation 6
            [0, 0, 0, 0, 0, y_hat, -1, 0],  # Equation 7
            [0, 0, 0, 0, -1, -x_hat, 0, 0], # Equation 8
        ])
        return x_hat, y_hat, A
    
    x_hat, y_hat, A = unit_vector (0,0)
    return A
    
def get_rhs(Pv, Ph) :
    Pv = 30 # assumed value
    Ph = 60 # assumed value

    # dimensions (Pv & Ph) are assumed to be any number

    # defining (b & h) again
    b = 10
    h = 15

    B = np.array([0.5*b*Pv, -(0.5*b*Pv + h*Ph), 0, 0, 0, 0, 0,-Ph])
    return B
```

## Assessment

Define Python variables
`b`, `h`, `Pv`, and `Ph`
for storing the parameters
$b$, $h$, $P_{\text{v}}$, and $P_{\text{h}}$,
respectively.

Use `get_coeffs()` and `get_rhs()`
to generate the coefficient matrix and the right-hand-side vector,
respectively storing them in Python variables
`coeffs` and `rhs`.

Use [`numpy.linalg.solve()`](https://numpy.org/doc/2.3/reference/generated/numpy.linalg.solve.html)
to solve for $\boldsymbol{x}$,
storing it to a Python variable `vars`.

```python
# ---------------------------
# Main: define parameters
# ---------------------------
b  = 10.0   # horizontal member length (assumed value)
h  = 15.0   # vertical height (assumed value)
Pv = 30.0   # vertical point load (assumed value)
Ph = 60.0   # horizontal point load (assumed value)

# The values of the variables (b,h,Pv,Ph) is just an assumption, since there is no specific given values for each dimension & point loads)


A = get_coeffs(b, h)
B = get_rhs (Pv, Ph)

print (A)
print ("")
print ("")
print ("")
print (B)
```

```python
x = np.linalg.solve(A,B)

print (x)
```

```python
# printing the unknown variales differently
# Horizontal & Vertical force of pin ( Cx & Cy ), Vertical Force of roller (Ay), Force Members ( AB, AD, BD, CD, BC)

print("Horizontal Force of the pin (Cx) is :",x[0],"N (Newton)")
print("Vertical Force of the pin (Cy) is :",x[1],"N (Newton)")
print("Vertical Force of the roller (Ay) is :",x[2],"N (Newton)")
print("Force Member (AB) is :",x[3],"N (Newton)")
print("Force Member (AD) is :",x[4],"N (Newton)")
print("Force Member (BD) is :",x[5],"N (Newton)")
print("Force Member (CD) is :",x[6],"N (Newton)")
print("Force Member (BC) is :",x[7],"N (Newton)")
```
